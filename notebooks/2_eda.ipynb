```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the final merged data
df = pd.read_csv('../data/processed/final_merged_data.csv')

# Set up styling
sns.set(style="whitegrid")

# 1. Target Variable Balance
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='is_late', palette='Set2')
plt.title('Late vs On-Time Deliveries')
plt.xticks([0,1], ['On-Time', 'Late'])
plt.ylabel('Count')
plt.show()

# 2. Average Shipping Days vs Delayed Status
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='is_late', y='shipping_days', palette='Set3')
plt.title('Shipping Days by Delivery Status')
plt.xticks([0,1], ['On-Time', 'Late'])
plt.show()

# 3. Most Common Product Categories (Top 10)
plt.figure(figsize=(10,4))
top_categories = df['product_category_name'].value_counts().head(10)
sns.barplot(x=top_categories.index, y=top_categories.values, palette='coolwarm')
plt.title('Top 10 Product Categories')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()

# 4. Delay Rate by Product Category (Top 10)
delay_rate = df.groupby('product_category_name')['is_late'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=delay_rate.index, y=delay_rate.values, palette='magma')
plt.title('Delay Rate by Product Category (Top 10)')
plt.ylabel('Delay Rate')
plt.xticks(rotation=45)
plt.show()

# 5. Delay Rate by Customer State (Top 10)
state_delay = df.groupby('customer_state')['is_late'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,4))
sns.barplot(x=state_delay.index, y=state_delay.values, palette='viridis')
plt.title('Delay Rate by Customer State (Top 10)')
plt.ylabel('Delay Rate')
plt.xticks(rotation=45)
plt.show()
```
